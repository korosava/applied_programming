```python
# тестові дані
z = 10
test_arr = [30, 12, 13, 100, 2, 3, 0]
```

```python
# (ЗАВДАННЯ_2)
# функція, що повертає значення найблище до заданого


def close(z, arr):
    '''
    функція, що повертає значення з масиву arr найблище до z
    '''

    temp = np.absolute(np.subtract(z, arr))
    index = np.argmin(temp)
    res = arr[index]    
    return res

print(close(z, test_arr))
```
